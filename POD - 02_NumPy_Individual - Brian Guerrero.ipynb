# Operación vectorizada vs. Operación loopeada
`numpy` no es solo poderoso por la gama de operaciones que podemos hacer en pocas lineas de código. Sino que por su eficiencia.

Supongamos que tenemos la función $f(x)=10*(x^2/e^x)$. Evaluar la función (en celdas apartes) entre -1 y 1 con 100,000 valores (i) con y (ii) sin un loop `for`. Medir tiempos de cada celda y sacar conclusiones.

```python

import time

def f_vectorizada(x):
    return 10 * (x**2 / np.exp(2))


x_values = np.linspace(-1, 1, 100000)

start_time = time.time()
result_vectorized = f_vectorizada(x_values)
end_time = time.time()

print(f"Tiempo con vectorización: {end_time - start_time:.6f} segundos")
```

```python
import math
def f_no_vectorizada(x):
    return 10 * (x**2 / math.exp(2))

start_time = time.time()
result_loop = [f_no_vectorizada(x) for x in x_values]
end_time = time.time()

print(f"Tiempo con loop: {end_time - start_time:.6f} segundos")
```

El resultado de la celda anterior (aprovechar lo obtenido), graficarlo como gráfico de lineas.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(x_values, result_vectorized, label="Vectorizada", alpha=0.8)
plt.plot(x_values, result_loop, label="Loop", linestyle='--', alpha=0.8)
plt.title("Comparación de evaluación: Vectorizada vs Loop")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

Por último, vamos a proceder a "binarizar" la imagen. Es decir, vamos a setear en 1 TODOS los pixeles donde la intensidad (es decir, el valor del pixel) sea mayor a 200. El resto, lo seteamos a 0. Mostrar el resultado.

Lea el archivo csv (sin `pandas` pero se puede usar `csv`) y quedese con todas las columnas que son numericas. Inspeccionarlo puede hacer el trabajo bastante facil.

Nota: Aunque `numpy` tenga una libreria para leer archivos, en este caso no no es útil por que todas las columas constan de diferentes tipo de dato y ademas el encabezado es `string`.

```python
import csv

with open('Boston.csv', 'r') as file:
    reader = csv.reader(file)
    header = next(reader)
    rows = [row for row in reader]
for row in rows[:5]:
    print(row)

numerical_data = []
for row in rows:
    numerical_row = []
    for value in row:
        try:
            numerical_row.append(float(value))
        except ValueError:
            pass
    numerical_data.append(numerical_row)

for row in numerical_data[:5]:
    print(row)
```

Una vez procesado el archivo csv, proceda a convertirlo en una matriz de `numpy`.

```python
numerical_data = []
with open('Boston.csv', 'r') as file:
    reader = csv.reader(file)
    header = next(reader)
    for row in reader:
        numerical_row = []
        for value in row:
            try:
                numerical_row.append(float(value))
            except ValueError:
                pass
        numerical_data.append(numerical_row)


numpy_matrix = np.array(numerical_data)


print("Matriz numpy:")
print(numpy_matrix)
print("\nDimensiones de la matriz:", numpy_matrix.shape)
```

Reporte el promedio de todas las columnas.

```python
column_means = numpy_matrix.mean(axis=0)

print("Promedio de todas las columnas:")
print(column_means)
```

Reporte la división entre una columna y otra (a elección cuales)

```python
columna_1 = numpy_matrix[:, 0]
columna_2 = numpy_matrix[:, 1]

division_columnas = columna_1 / columna_2

print("División entre la columna 1 y la columna 2:")
print(division_columnas)
```
